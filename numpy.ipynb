```python
np.arange(0, 10) # Creates an array of values from 0 to 9
```

```python
np.linspace(0, 10, 5) # Creates an array of 5 values evenly spaced between 0 and 10
```

```python
np.random.rand(3,3) # Creates a 3x3 array of random values between 0 and 1
```

```python
np.random.randint(1,100, (5,5)) # Creates a 5x5 array of random integers between 1 and 100
```

```python
np.sum(x)                     # Sum of all elements in the array
```

```python
np.mean(y, axis=1)
```

```python
np.amax(y)
```

```python
np.amax(y, axis=0)  # Max of each column
```

```python
np.amax(y, axis=1)  # Max of each row
```
